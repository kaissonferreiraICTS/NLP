## 1.1 build functions

```python
# Remove as tags de HTML
def _remove_tags_html(text):
    text = re.sub('<[^<]+?>', '', str(text))
    return text
```

```python
# Remove web sites:
def _remove_web(text):
    text = re.sub(r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]|\(([^\s()<>]|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?]))",
                "xxxWEBxxx",
                str(text)
            )
    text = ' '.join([w for w in text.split() if '.com' not in w])
    return text
```

```python
# Remove e-mail
def _remove_email(text):
    text = ' '.join([w for w in text.split() if '@' not in w])
    return text
```

```python
# Remove os Emojis
def _remove_emoji(string):
    emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
                "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
def _remove_big_term(text):
    word_size = 35
    word_replace = 'xxxBIGTERMxxx'
    text = ' '.join([w if len(w) <= word_size else word_replace for w in text.split(' ')])
    return text
```

Função para deixar o texto sem acento e apenas com letra minúsculas.

```python
def _sub_char_especial_min(text):
    # Convertendo texto para letras minúsculas
    text = text.lower()

    # Substitui caracteres especiais por letras sem acentos

    text = re.sub(r"[áãàâ]", "a", text)
    text = re.sub(r"[éêè]", "e", text)
    text = re.sub(r"[íîì]", "i", text)
    text = re.sub(r"[óõôò]", "o", text)
    text = re.sub(r"[úùû]", "u", text)
    text = re.sub(r"[ç]", "c", text)
    text = re.sub(r"&aacute;", "a", text)
    text = re.sub(r"&atilde;", "a", text)
    text = re.sub(r"&Atilde;","a", text) # Letra maiúscula
    text = re.sub(r"&agrave;", "a", text)
    text = re.sub(r"&acirc;", "a", text)
    text = re.sub(r"&aring;", "a", text)
    text = re.sub(r"&auml;", "a", text)
    text = re.sub(r"&aelig;", "ae", text)
    text = re.sub(r"&eacute;", "e", text)
    text = re.sub(r"&ecirc;", "e", text)
    text = re.sub(r"&egrave;", "e", text)
    text = re.sub(r"&euml;", "e", text)
    text = re.sub(r"&iacute;", "i", text)
    text = re.sub(r"&icirc;", "i", text)
    text = re.sub(r"&iuml;", "i", text)
    text = re.sub(r"&oacute;", "o", text)
    text = re.sub(r"&otilde;", "o", text)
    text = re.sub(r"&ocirc;", "o", text)
    text = re.sub(r"&ograve;", "o", text)
    text = re.sub(r"&uacute;", "u", text)
    text = re.sub(r"&ucirc;", "u", text)
    text = re.sub(r"&ugrave;", "u", text)
    text = re.sub(r"&uuml;", "u", text)
    text = re.sub(r"&ccedil;", "c", text)
    text = re.sub(r"&ntilde;", "n", text)
    text = re.sub(r"&lt;", "<", text)
    text = re.sub(r"&gt;", ">", text)
    text = re.sub(r"&amp;", "&", text)
    text = re.sub(r"&reg;", "&", text) #any special char to be replaced
    text = re.sub(r"&copy;", "&", text) #any special char to be replaced
    text = re.sub(r"&quot;","'", text)
    text = re.sub(r"&lsquo;","'", text)
    text = re.sub(r"&rsquo;","'", text)
    return text
```

Função para deixar o texto com acento e letras maiúsculas e minúsculas.

```python
def _sub_char_especial_normal(text):
    # Substitui caracteres especiais por letras com acentos

    # Letra minúsculas
    text = re.sub(r"&aacute;", "á", text)
    text = re.sub(r"&atilde;", "ã", text)
    text = re.sub(r"&agrave;", "à", text)
    text = re.sub(r"&acirc;", "â", text)
    text = re.sub(r"&aring;", "a", text)
    text = re.sub(r"&auml;", "a", text)
    text = re.sub(r"&aelig;", "ae", text)
    text = re.sub(r"&eacute;", "é", text)
    text = re.sub(r"&ecirc;", "ê", text)
    text = re.sub(r"&egrave;", "è", text)
    text = re.sub(r"&euml;", "e", text)
    text = re.sub(r"&iacute;", "í", text)
    text = re.sub(r"&icirc;", "î", text)
    text = re.sub(r"&iuml;", "i", text)
    text = re.sub(r"&oacute;", "ó", text)
    text = re.sub(r"&otilde;", "õ", text)
    text = re.sub(r"&ocirc;", "ô", text)
    text = re.sub(r"&ograve;", "ò", text)
    text = re.sub(r"&uacute;", "ú", text)
    text = re.sub(r"&ucirc;", "û", text)
    text = re.sub(r"&ugrave;", "ù", text)
    text = re.sub(r"&uuml;", "u", text)
    text = re.sub(r"&ccedil;", "ç", text)
    text = re.sub(r"&ntilde;", "não", text)

    # Letras maiúsculas
    text = re.sub(r"&Aacute;", "Á", text)
    text = re.sub(r"&Atilde;","Ã", text) # Letra maiúscula
    text = re.sub(r"&Agrave;", "À", text)
    text = re.sub(r"&Acirc;", "Â", text)
    text = re.sub(r"&Aring;", "A", text)
    text = re.sub(r"&Auml;", "A", text)
    text = re.sub(r"&Aelig;", "AE", text)
    text = re.sub(r"&Eacute;", "É", text)
    text = re.sub(r"&Ecirc;", "Ê", text)
    text = re.sub(r"&Egrave;", "È", text)
    text = re.sub(r"&Euml;", "E", text)
    text = re.sub(r"&Iacute;", "Í", text)
    text = re.sub(r"&Icirc;", "Î", text)
    text = re.sub(r"&Iuml;", "I", text)
    text = re.sub(r"&Oacute;", "Ó", text)
    text = re.sub(r"&Otilde;", "Õ", text)
    text = re.sub(r"&Ocirc;", "Ô", text)
    text = re.sub(r"&Ograve;", "Ò", text)
    text = re.sub(r"&Uacute;", "Ú", text)
    text = re.sub(r"&Ucirc;", "Û", text)
    text = re.sub(r"&Ugrave;", "Ù", text)
    text = re.sub(r"&Uuml;", "U", text)
    text = re.sub(r"&Ccedil;", "Ç", text)
    text = re.sub(r"&Ntilde;", "Não", text)
    text = re.sub(r"&Yacute;", "Y", text)

    #any special char to be replaced
    text = re.sub(r"&lt;", "<", text)
    text = re.sub(r"&gt;", ">", text)
    text = re.sub(r"&amp;", "&", text)
    text = re.sub(r"&reg;", "&", text) 
    text = re.sub(r"&copy;", "&", text) 
    text = re.sub(r"&quot;","'", text)
    text = re.sub(r"&lsquo;","'", text)
    text = re.sub(r"&rsquo;","'", text)
    return text
```

```python
# Termos em code HTML
def _remove_char_html(text):
    text = re.sub(r"&ndash;","", text)
    text = re.sub(r"&ordf;","", text)
    text = re.sub(r"&nbsp;","", text)
    text = re.sub(r"&cent;","", text)
    text = re.sub(r"&pound;","", text)
    text = re.sub(r"&sect;","", text)
    text = re.sub(r"&copy;","", text)
    text = re.sub(r"&laquo;","", text)
    text = re.sub(r"&raquo;","", text)
    text = re.sub(r"&reg;","", text)
    text = re.sub(r"&deg;","", text)
    text = re.sub(r"&plusmn;","", text)
    text = re.sub(r"&para;","", text)
    text = re.sub(r"&middot;","", text)
    text = re.sub(r"&frac12;","", text)
    text = re.sub(r"&mdash;","", text)
    text = re.sub(r"&lsquo;","", text)
    text = re.sub(r"&rsquo;","", text)
    text = re.sub(r"&sbquo;","", text)
    text = re.sub(r"&ldquo;","", text)
    text = re.sub(r"&rdquo;","", text)
    text = re.sub(r"&bdquo;","", text)
    text = re.sub(r"&dagger;","", text)
    text = re.sub(r"&Dagger;","", text)
    text = re.sub(r"&bull;","", text)
    text = re.sub(r"&hellip;","", text)
    text = re.sub(r"&prime;","", text)
    text = re.sub(r"&Prime;","", text)
    text = re.sub(r"&euro;","", text)
    text = re.sub(r"&trade;","", text)
    text = re.sub(r"&asymp;","", text)
    text = re.sub(r"&ne;","", text)
    text = re.sub(r"&le;","", text)
    text = re.sub(r"&ge;","", text)
    text = re.sub(r"&lt;","", text)
    text = re.sub(r"&gt;","", text)
    return text
```

```python
def transform(text):
    text = _remove_tags_html(text)
    text = _remove_web(text)
    text = _remove_email(text)
    text = _remove_emoji(text)
    text = _remove_big_term(text)
    text = _sub_char_especial_normal(text)
    text = _remove_char_html(text)
    return text

# Transformado tudo em minúsculo e sem acentuação.
def transform_lower(text):
    text = _remove_tags_html(text)
    text = _remove_web(text)
    text = _remove_email(text)
    text = _remove_emoji(text)
    text = _remove_big_term(text)
    text = _sub_char_especial_min(text)
    text = _remove_char_html(text)
    return text
```

## 1.4 Apply functions

* Aplicando uma função que compila todas as outras funções auxiliares na coluna relato da base de dados 'base_v1'.

```python
df1['Relato_tratado'] = df1['Relato'].map(transform_lower)
print('Rows: {} \nColumns: {}'.format(df1.shape[0],df1.shape[1]))
df1.head()
```
